```python
from tqdm import tqdm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
from src.training.align_images import AlignImages

from src.training.load_config import load_config
```

```python
run_config = load_config()
path = Path(run_config['jpeg_output_subdir'])
outpath = Path(run_config['base_data_dir']) / run_config['jpeg_output_subdir']
secondary_align_csv = outpath / run_config['secondary_align_csv']
```

```python
dataset = AlignImages(outpath, run_config['align_csv'])
```

```python
metric_list = []
for i in tqdm(range(len(dataset))):
    gt, noisy, aligned, metrics = dataset[i]
    metric_list.append(metrics)
```

```python

df = pd.DataFrame(metric_list)
df.to_csv(secondary_align_csv)
```

```python
df = pd.read_csv(secondary_align_csv)
```
